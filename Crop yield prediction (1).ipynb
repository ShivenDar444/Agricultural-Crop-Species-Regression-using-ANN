```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
df = pd.read_csv('Dataset/yield_df.csv')
df.head()
```

```python
df.drop(['Unnamed: 0','Area','Year'],axis=1,inplace=True)
df.head()
```

```python
df = pd.get_dummies(df,columns=['Item'],drop_first=True)
df.head()
```

```python
y = df['hg/ha_yield']
X = df.drop(['hg/ha_yield'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
ann = tf.keras.models.Sequential()  # initializing the ANN
ann.add(tf.keras.layers.Dense(128,input_dim=12,activation='relu'))  # 1st hidden layer
ann.add(tf.keras.layers.Dense(64,activation='relu'))  # 2nd hidden layer
ann.add(tf.keras.layers.Dense(1,activation='linear'))  # output layer
```

```python
# ann.compile(optimizer='adam',loss='mean_squared_error')
ann.compile(optimizer='adam',loss='mean_squared_error')
ann.fit(X_train,y_train,batch_size=16,epochs=500)
```

```python
y_pred = ann.predict(X_test)

from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
